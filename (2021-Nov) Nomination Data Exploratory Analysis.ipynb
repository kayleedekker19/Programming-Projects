```python
#set seed
set.seed(11)
#number of cross-validations is 5, cp value 0.000007
#choosen attribute are the same as lm1
rt.all <- rpart(ObamaRate ~ Black+Hispanic+Asian+AmericanIndian+Hawaiian+White+MedianIncome+AverageIncome+
          MedicareRate+UnemployRate+IncomeAbove75K+SocialSecurityRate+
          RetiredWorkers+Homeowner+Poverty,
        data = elect.df.known,
        xval=5, #number of cross-validations
        cp = 0.000007) 
```

```python
plotcp(rt.all,upper = "splits")
```

```python
#determines the optimal cp corresponding to the tree with the smallest number of splits
optimalCP <- function(rt.model){
    df<-as.data.frame(rt.model$cptable)
    minerr <- min(df[,"xerror"])
    minerr.xstd <- df[df$xerror==minerr,"xstd"]
    df[df$xerror<minerr+minerr.xstd,][1,"CP"]}

(best_cp <- optimalCP(rt.all))
```

```python
#"prune" the optimal CP value back
rt.all.opt <- prune(rt.all, cp=best_cp)
```

```python
#decision tree
rpart.plot(rt.all.opt, type = 1, extra = 1)
```

```python
# load packages.
library(lattice)
library(ggplot2)
library(caret)
library(zoo) 
library("forecast")
# load the US_Births data
US_Births <-read.csv("US_Births.csv")
```

```python
class(US_Births$Yr_Mo)
# transform the numeric to a date format 
as.Date <- as.Date(paste0(as.character(US_Births$Yr_Mo), '01'), format='%Y%m%d')
class(as.Date) # check it is now indeed a date

# extract months and years from as.Date
Month <-format(as.Date, "%m")
Year <- format(as.Date, "%Y")

# adding months and years to US_Births dataframe using cbind()
US_Births_1 <- cbind(US_Births, Month, Year)
```

```python
# prepare the data as time-series objects
US_Births.ts <- ts(US_Births_1$Live.Births, start = c(2007, 01),
                end = c(2012, 06), freq = 12)

# We first use this optional code from repr package to change the default size of output plots
library(repr)
options(repr.plot.width=15, repr.plot.height=6) # change plot size # Make plot
plot(US_Births.ts, ylab = "Live.Births")
```

```python
# Generate forecast model
# 1. smoothing method, based on averaging over multiple
# observations to reduce noice
# Centered moving averages, 3-month window
Live.Births.ma_c=rollmean(US_Births.ts,3,align='center') 
plot(US_Births.ts, ylab = "Live.Births",col='blue') 
lines(Live.Births.ma_c,col='red')
```

```python
# Holt's linear trend model, ideal for time series data trending
# up/down, or varying seasonally 
# trend model, AAN
(US_Births.ets.AAN <- ets(US_Births.ts, model = "AAN"))

#Check RMSE 
rmse.ets <- function (etsmodel) sqrt(etsmodel$mse) 
(error <- rmse.ets(US_Births.ets.AAN))

# save RMSE 
#Save RMSE
US_Births.rmse=data.frame(RMSE=error, Model='AAN')
#US_Births.rmse=rbind(US_Births.rmse, data.frame(RMSE=error, Model='ANN')) 
US_Births.rmse

# Plot the decomposition of fitted model showing both actual values
# (observed) and the exponential smoothed average(level)
plot(US_Births.ets.AAN)
# slope(b) value indicates the trend change in the fitted value per unit of time
# (eg per month)

# Generate forecast to February 2013, there are 8 steps ahead 
US_Births.ets.AAN.pred <- forecast(US_Births.ets.AAN, h = 8) 
plot(US_Births.ets.AAN.pred)
```

```python
# seasonality model AAA
(US_Births.ets.AAA <- ets(US_Births.ts, model = "AAA"))

#Check RMSE 
#rmse.ets <- function (etsmodel) sqrt(etsmodel$mse) 
(error <- rmse.ets(US_Births.ets.AAA))

# save RMSE 
#Save RMSE
US_Births.rmse=rbind(US_Births.rmse, data.frame(RMSE=error, Model='AAA')) 
US_Births.rmse

# Plot the decomposition of fitted model showing both actual values
# (observed) and the exponential smoothed average(level)
plot(US_Births.ets.AAA)
# slope(b) value indicates the trend change in the fitted value per unit of time
# (eg per month), season(s) indicates repeating short term cycle of the time series

# Generate forecast to February 2013, there are 8 steps ahead 
# prediction at 80% confidence level
US_Births.ets.AAA.pred <- forecast(US_Births.ets.AAA, h = 8,
                                prediction.interval = TRUE, level= 0.8) 
plot(US_Births.ets.AAA.pred)
```

```python
# Compare the two models & print the final US Births forecasts for Feb 2013 
US_Births.ets.AAN.pred$mean[8] 
US_Births.ets.AAA.pred$mean[8]
```
